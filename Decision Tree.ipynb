```python
                                                                                                                                                                                                            
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

col_names = ['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe','Type']

df=pd.read_csv("glass.csv", header=None,names=col_names)
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python

feature = ['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']
x=df[feature]
# y is target values
y=df.Type

# Split dataset into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1) # 70% training and 30% test

print("Original data x:",x.shape)
print("After spliting x_traing shape:",np.shape(x_train))
print("After spliting x_test size:",np.shape(x_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="gini",max_depth=5,random_state=0)
# Train Decision Tree Classifer
clf = clf.fit(x_train,y_train)
#Predict the response for test dataset
y_pred = clf.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("Missclassified :",(y_test!=y_pred).sum())
print("Accuracy:",accuracy_score(y_test, y_pred))

```

### Find Best Accuracy

```python
max_dep={}# find best accuracy with max_depth value
miss=[]  #find minimum missclassified 
for i in range(1,101):
    clf = DecisionTreeClassifier(criterion="gini",max_depth=i)
    clf = clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    b=(y_test!=y_pred).sum()
    miss.append(b)
    a=accuracy_score(y_test, y_pred)
    max_dep[i]=a

print("\nMissclassified: ",min(miss))
print("Best max_depth value for max Aaccuracy: ",max(max_dep,key=max_dep.get))    
print("Accuracy: ",max_dep[max(max_dep,key=max_dep.get)])    
```
