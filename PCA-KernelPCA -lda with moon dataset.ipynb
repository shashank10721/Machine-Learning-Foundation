```python
from sklearn.datasets import make_moons
X,y=make_moons(n_samples=200,noise=0.05,random_state=0)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(X[y==0,0],X[y==0,1],color="red",marker="^",label='Class-1')
plt.scatter(X[y==1,0],X[y==1,1],color="green",marker="o",label='Class-2')
plt.legend()
plt.title("Moon Dataset")
plt.grid()
plt.xlabel("F1")
plt.ylabel("F2")
plt.show()
```

```python

from sklearn.decomposition import PCA
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X)
plt.figure()
plt.scatter(X_pca[y==0,0],X_pca[y==0,1],color="red",marker="^",label='Class-1')
plt.scatter(X_pca[y==1,0],X_pca[y==1,1],color="blue",marker="o",label='Class-2')
plt.legend()
plt.title("Moon Dataset With PCA")
plt.grid()
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.show()
```

```python
from sklearn.decomposition import KernelPCA
kpca=KernelPCA(n_components=2,kernel="rbf",gamma=20)
X_kpca=kpca.fit_transform(X)
plt.figure()
plt.scatter(X_kpca[y==0,0],X_kpca[y==0,1],color="red",marker="^",label='Class-1')
plt.scatter(X_kpca[y==1,0],X_kpca[y==1,1],color="orange",marker="o",label='Class-2')
plt.legend()
plt.title("Moon Dataset With  KernelPCA")
plt.grid()
plt.xlabel("KPCA1")
plt.ylabel("KPCA2")
plt.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda=LDA(n_components=2) #two pca
X_lda=lda.fit_transform(X,y)
plt.figure()
plt.scatter(X_lda[y==0],X_lda[y==0],color="red",marker="^",label='Class-1')
plt.scatter(X_lda[y==1],X_lda[y==1],color="orange",marker="o",label='Class-2')
plt.legend()
plt.title("Moon Dataset With LDA")
plt.grid()
plt.xlabel("LDA1")
plt.ylabel("LDA2")
plt.show()
```
